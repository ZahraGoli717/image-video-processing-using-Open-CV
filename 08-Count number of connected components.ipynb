## üß©<span style='color:#4B0082'> What ‚ÄúConnected Components‚Äù Means</span>
 
A connected component is a group of pixels in a binary image that are connected to each other (typically by 4 or 8-neighbor connectivity) and share the same value (usually 1 or white).
 
In simpler terms ‚Äî if you have blobs or objects in a binary image, each distinct blob is a connected component.
 
 ---
 
### üéØ Where Connected Component Counting Is Useful
 
Here are some practical applications across different fields:
 
**1. Object Counting**
 
Counting coins, pills, or cells in X-ray/MRI/microscope images.
 
Example: In manufacturing, you can count defective parts or missing components on an assembly line.
 
 
**2. Document Image Analysis**
 
Detecting and separating characters or words in scanned documents.
 
Useful for Optical Character Recognition (OCR) preprocessing.
 
 
**4. Traffic & Surveillance**
 
Counting vehicles, pedestrians, or other moving objects after segmentation.
 
 
**5. Agriculture**
 
Counting fruits, seeds, or plants from aerial or field images for yield estimation.
 
 
**6. Quality Inspection**
 
Detecting cracks, voids, or missing parts in industrial quality control.
 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image_path = 'number threshold.jpg'  # Replace with your image path
img = cv2.imread(image_path)
```

```python
plt.imshow(img[...,::-1])
```

```python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply a large Gaussian blur to reduce noise
kernel_size = (13, 13)
blurred = cv2.GaussianBlur(gray, kernel_size, 0)

# Use adaptive thresholding to create a binary image
binary = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 9, 2)

plt.imshow(binary[...,::-1])
```

```python
# Find connected components
num_labels, labels = cv2.connectedComponents(binary)

# Print the number of connected components found (excluding background)
print(f"Number of connected components found: {num_labels - 1}")
```
